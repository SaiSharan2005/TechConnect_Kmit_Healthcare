```python
import pandas as pd
data = pd.read_csv('corona_tested_individuals_ver_0083.csv')
data
```

```python
from pyasn1.type.univ import Null
data_edit = data.drop("test_date",axis=1)
# data_edit['corona_result']!='other'
data_edit = data_edit[data_edit['corona_result']!='other']
# data_edit = data_edit[data_edit['age_60_and_above']!='']
data_edit = data_edit.dropna()
data_edit.tail()
data_edit.head(60)
```

```python
data_edit['corona_result']=data_edit['corona_result'].map({'negative':0,'positive':1})
data_edit['age_60_and_above']=data_edit['age_60_and_above'].map({'No':0,'Yes':1})
data_edit['gender']=data_edit['gender'].map({'female':0,'male':1})
data_edit['test_indication']=data_edit['test_indication'].map({'Other':0,'Contact with confirmed':1,'Abroad':2})
dat_edit= data_edit.dropna()
print(data_edit.size)
data_edit.isnull().sum()
```

```python

X= data_edit.drop('corona_result',axis=1)
Y= data_edit.iloc[:,5:6]
# X = X.dropna()
# Y = Y.dropna() 
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l1','l2','elasticnet'],'C':[1,2,3,4,5,6,10,20,30,40,50],'max_iter':[100,200,300]}
```

```python
classifier_regressor=GridSearchCV(classifier,param_grid=parameter,scoring='accuracy',cv=5)
```

```python
classifier.fit(X_train,Y_train)
```

```python
Y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(Y_pred,Y_test)
print(score)
```
